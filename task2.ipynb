```python
def largest(a,b,c):
    l=a
    if b>l:
        l=b 
    if c>l: 
        l=c
    return l

print(largest(22,85,25))
```

```python
def prime(n):
    if n<=1:
        print("Not a prime number")
        return 
    for i in range(2,n//2+1):
        if n%i==0:
            print("Not a prime number")
            return
    print("Prime number")   

    

prime(40)
prime(25)
```

```python
def fib(n):
    if n<=0:
        return "invalid"
    elif n==1:
        return 0    
    elif n==2:
        return 1
    else:
        return fib(n-1)+fib(n-2)    

fib(30)
```
